```python
import cv2 as cv
from utils.io_utils import get_cam_images, output_calib_results, draw_detections
from utils.calib_utils import detect_keypoints, to_lists
```

## Basler0

```python
pattern_size = (8, 6)
data_path0 = "/home/konstantin/datasets/bandeja-sequence/2020-11-17-18-52-32-chess/camera0"
```

```python
images0 = get_cam_images(data_path0, 4)
```

```python
len(images0)
```

```python
results0 = detect_keypoints(images0, pattern_size)
```

```python
draw_detections(images0, results0, pattern_size, 0)
```

```python
criteria = (cv.TERM_CRITERIA_EPS + cv.TERM_CRITERIA_MAX_ITER, 30, 0.0001)

scene_points0, loc_kp0 = to_lists(results0)

shape = images0[0].shape[::-1][1:]

# Note: K3 dist. coeff. is set to zero, because ros requires 4 distortion parameters. Change if needed.
reproj_error0, intrinsics0, dist_coeff0, _, _ = cv.calibrateCamera(scene_points0, loc_kp0, shape, None, None, flags=cv.CALIB_FIX_K3, criteria=criteria)

print(reproj_error0)
```

```python
output_calib_results(intrinsics0, dist_coeff0, shape, images0, 0, 'output/basler0_intrinsincs')
```
